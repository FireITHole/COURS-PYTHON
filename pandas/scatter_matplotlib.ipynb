```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_athletes = pd.read_csv('data/bios.csv')
```

```python
df_athletes.info()
```

```python
df_athletes.dropna(subset=['height_cm', 'weight_kg', 'born_date'], inplace=True)
```

```python
df_athletes.info()
```

```python
df_athletes["born_year"] = pd.to_datetime(df_athletes["born_date"]).dt.year
```

```python
df_mens = df_athletes[['height_cm', 'weight_kg', 'born_year']]
```

```python
df_mens.sort_values(by=['height_cm', 'weight_kg'], inplace=True)
```

```python
df_mens
```

```python
plt.scatter(df_mens['weight_kg'], df_mens['height_cm'], color="orange", alpha=0.2, s=20)
plt.ylabel('Height (cm)')
plt.xlabel('Weight (kg)')
plt.title('Height vs Weight of Olympic Athletes')
```

```python
plt.scatter(x='weight_kg', y='height_cm', data=df_mens, c='born_year', cmap='rainbow', alpha=.5, s=15, vmin=df_mens["born_year"].min(), vmax=df_mens["born_year"].max())
plt.ylabel('Height (cm)')
plt.xlabel('Weight (kg)')
plt.title('Height vs Weight of Olympic Athletes')
plt.colorbar(label='Born Year')
plt.grid(linestyle='--', alpha=0.5)
```
