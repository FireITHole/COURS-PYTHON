```python
import pandas as pd
```

```python
netflix_data = pd.read_csv('data/netflix_titles.csv')
```

```python
netflix_data.drop(columns=['director'], inplace=True)
```

```python
netflix_data.head()
```

```python
netflix_data.shape
```

```python
netflix_data.index
```

```python
netflix_data.dropna(subset=['cast', 'country', 'date_added'], inplace=True)
```

```python
netflix_data.shape
```

```python
netflix_data.isna().sum().sort_values(ascending=False)
```

```python
most_seen_rating = netflix_data.rating.mode()
```

```python
netflix_data['rating'].fillna(most_seen_rating[0], inplace=True)
```

```python
netflix_data.isna().sum().sort_values(ascending=False)
```

```python
netflix_data['type'].unique()
```

```python
netflix_data["is_movie"] = netflix_data["type"].apply(lambda t: True if t == "Movie" else False)
```

```python
netflix_data["movie_duration"] = netflix_data[netflix_data["is_movie"] == True]["duration"].str.split(expand=True)[0].astype(int)
```

```python
netflix_data["nb_seasons_show"] = netflix_data[netflix_data["is_movie"] == False]["duration"].str.split(expand=True)[0].astype(int)
```

```python
netflix_data.drop(columns=["duration", "type"], inplace=True)
```

```python
netflix_data.date_added = pd.to_datetime(netflix_data.date_added.str.strip(), format="%B %d, %Y")
```

```python
netflix_data
```

```python
netflix_data.index = netflix_data["show_id"]
```

```python
netflix_data.drop(columns="show_id", inplace=True)
```

```python
netflix_data
```

```python
release_year_x_movie_duration_median = netflix_data[netflix_data["is_movie"] == True].groupby("release_year")["movie_duration"].median()
release_year_x_movie_duration_median
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(15, 7))
release_year_x_movie_duration_median.plot(kind="line")
plt.xlabel("Année de sortie")
plt.ylabel("Durée médiane des films")
plt.yticks(np.arange(0, release_year_x_movie_duration_median.max()+1, 20))
plt.xticks(np.arange(release_year_x_movie_duration_median.index.min(), release_year_x_movie_duration_median.index.max()+1, 5), rotation=45)
plt.grid()
```

```python
nb_films_x_annee = netflix_data[netflix_data["is_movie"] == True].groupby("release_year")["is_movie"].sum()
nb_films_x_annee
```

```python
plt.figure(figsize=(15, 7))
release_year_x_movie_duration_median.plot(kind="line")
plt.xlabel("Année de sortie")
plt.ylabel("Durée médiane des films")
plt.yticks(np.arange(0, release_year_x_movie_duration_median.max()+1, 20))
plt.xticks(np.arange(release_year_x_movie_duration_median.index.min(), release_year_x_movie_duration_median.index.max()+1, 5), rotation=45)
plt.grid()

for year, nb in tuple(nb_films_x_annee.items())[::5]:
    plt.text(year, release_year_x_movie_duration_median[year]+5, f"{nb} films", fontsize=8, color="red")
```

```python
df_movies_countries = netflix_data[netflix_data["is_movie"] == True]["country"].str.strip().str.split(", ", expand=True).stack().reset_index(level=1, drop=True).reset_index()
df_movies_countries.columns = ["show_id", "country"]
```

```python
df_movies_countries["country"].value_counts(ascending=False).reset_index()
```

```python
plt.figure(figsize=(20, 7))
df_movies_countries["country"].value_counts(ascending=True).plot(kind="bar")
plt.xlabel("Pays")
plt.ylabel("Nombre de films")
plt.grid()
```

```python
nb_shows_movies_x_year = netflix_data.groupby("release_year")["is_movie"].value_counts().unstack().fillna(0)
nb_shows_movies_x_year.columns = ["nb_shows", "nb_movies"]
```

```python
nb_shows_movies_x_year = nb_shows_movies_x_year.reset_index()
nb_shows_movies_x_year
```

```python
nb_shows_movies_x_year.plot(figsize=(15, 7), kind="bar", stacked=True, x="release_year", y=["nb_shows", "nb_movies"], label=["Séries", "Films"])
plt.xlabel("Année de sortie")
plt.ylabel("Nombre de films/séries")
plt.xticks(np.arange(nb_shows_movies_x_year.index.min(), nb_shows_movies_x_year.index.max()+1, 5), rotation=45)
plt.grid()
```
